```python
import re

# 1. Load your raw unique array
storage_array = df['Memory']

# 2. Define the parser function using the custom Regex pattern
def engineer_storage_features(text):
    ssd, hdd, flash, hybrid = 0, 0, 0, 0
    
    # Standardise text to lowercase to ensure pattern matches flawlessly
    clean_text = str(text).lower()
    
    # Your universal pattern: (Number) (Unit) (Drive Type)
    pattern = r'(\d+(?:\.\d+)?)\s*(gb|tb)\s*(ssd|hdd|flash storage|hybrid)'
    matches = re.findall(pattern, clean_text)
    
    for size, unit, drive_type in matches:
        capacity = float(size)
        
        # Convert Terabytes to Gigabytes
        if unit == 'tb':
            capacity *= 1024
            
        # Accumulate numerical values into the correct hardware category
        if drive_type == 'ssd':
            ssd += capacity
        elif drive_type == 'hdd':
            hdd += capacity
        elif drive_type == 'flash storage':
            flash += capacity
        elif drive_type == 'hybrid':
            hybrid += capacity
            
    return int(ssd), int(hdd), int(flash), int(hybrid)

# 3. Apply the function over the array and structure into a clean DataFrame
cleaned_rows = []
for item in storage_array:
    ssd_gb, hdd_gb, flash_gb, hybrid_gb = engineer_storage_features(item)
    cleaned_rows.append({
        "MemoryStorage": item,
        "SSD_GB": ssd_gb,
        "HDD_GB": hdd_gb,
        "Flash_GB": flash_gb,
        "Hybrid_GB": hybrid_gb
    })

df_cleaned = pd.DataFrame(cleaned_rows)
df_cleaned
```

```python
import re
import numpy as np
import pandas as pd

# Raw input array
cpu_array = df['Cpu'].values

def extract_cpu_features_lean(text):
    text_str = str(text)
    
    # 1. Base Regex: Extract Brand, Middle text, and Clock Speed
    base_pattern = r'^(Intel|AMD|Samsung)\s+(.*?)\s+(\d+(?:\.\d+)?)\s*ghz'
    match = re.search(base_pattern, text_str, re.IGNORECASE)
    
    if not match:
        return "Unknown", "Other", 0.0
        
    brand = match.group(1)
    model_text = match.group(2).lower()
    speed = float(match.group(3))
    
    # 2. Extract Simplified Broad Category
    if 'core i7' in model_text:
        category = 'Intel Core i7'
    elif 'core i5' in model_text:
        category = 'Intel Core i5'
    elif 'core i3' in model_text:
        category = 'Intel Core i3'
    elif 'celeron' in model_text:
        category = 'Intel Celeron'
    elif 'pentium' in model_text:
        category = 'Intel Pentium'
    elif 'ryzen' in model_text:
        category = 'AMD Ryzen'
    elif 'a9-series' in model_text or 'a6-series' in model_text:
        category = 'AMD A-Series'
    elif 'e-series' in model_text:
        category = 'AMD E-Series'
    elif 'atom' in model_text:
        category = 'Intel Atom'
    elif 'xeon' in model_text:
        category = 'Intel Xeon'
    else:
        category = 'Other'

    return brand, category, speed

# Build the simplified feature matrix
rows = []
for item in cpu_array:
    brand, cat, speed = extract_cpu_features_lean(item)
    rows.append({
        "Raw_CPU_Text": item,
        "CPU_Brand": brand,
        "CPU_Category": cat,
        "Clock_Speed_GHz": speed
    })

df_cpu_lean = pd.DataFrame(rows)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
x=df.drop('Price_euros',axis=1)
y=np.log(df['Price_euros'])
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
xtrain.sample(5)
```

```python
trf = ColumnTransformer(
    transformers=[
        ('encode', OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore'), 
         ['Company', 'TypeName', 'Gpu', 'OpSys', 'CPU_Brand', 'CPU_Category'])
    ], 
    remainder='passthrough'
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error

algo = RandomForestRegressor(n_estimators=100, random_state=42)

pipe = Pipeline(steps=[
    ('step1', trf),   # Your ColumnTransformer from earlier
    ('step2', algo)   # The algorithm
])

pipe.fit(xtrain, ytrain)

y_pred = pipe.predict(xtest)

print('R2 Score:', r2_score(ytest, y_pred))
print('MAE:', mean_absolute_error(ytest, y_pred))
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(pipe, x, y, cv=5, scoring='r2')

print("R2 Scores for all 5 splits:", scores)
print("Average R2 Score:", scores.mean())
```
